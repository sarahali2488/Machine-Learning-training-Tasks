# Linear Regression: Building model from scratch

```
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```
df = pd.read_csv('/content/sample_data/Salary_dataset.csv')

df
```

```
X = df.iloc[:,1] #all rows with first column: year experience
y = df.iloc[:,2] #all rows with second column: salary
X
```

```
y
```

```
model = Sequential() #each layer passes its output to the second layer
model.add(Dense(1, input_dim=1)) #To build the model, dense means make one layer
model.add(Activation('linear')) #The activation func is linear as lin, not polynomial as curved line to fit data

model.summary()
```

```
model.compile(loss='mse',optimizer='sgd', metrics=['mse']) #passing model parameter
```

```
history = model.fit(X,y,epochs=200,verbose=1) #verbose shows details of the model
```

the loss is 53%. we can improve it

```
pred = model.predict(X)
pred
```

```
plt.scatter(X, y, c='blue') #draw the data, blue the data
plt.plot(X, pred, color='g') #green the model. if the green line fits the data correct so the model is perfect
plt.show()
```

# Building the model using sklearn model

```
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
# y = 1 * x_0 + 2 * x_1 + 3
y = np.dot(X, np.array([1, 2])) + 3

reg = LinearRegression().fit(X, y)
reg.score(X, y)
#reg.coef_
```
