# **Pandas Grouping and Aggregating Tasks:**

```python
"""
Test Data:
   school class            name date_Of_Birth   age  height  weight  address
S1   s001     V  Alberto Franco     15/05/2002   12    173      35  street1
S2   s002     V    Gino Mcneill     17/05/2002   12    192      32  street2
S3   s003    VI     Ryan Parkes     16/02/1999   13    186      33  street3
S4   s001    VI    Eesha Hinton     25/09/1998   13    167      30  street1
S5   s002     V    Gino Mcneill     11/05/2002   14    151      31  street2
S6   s004    VI    David Parkes     15/09/1997   12    159      32  street4
"""
import pandas as pd

data = {
    'school': ['s001', 's002', 's003', 's001', 's002', 's004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco', 'Gino Mcneill', 'Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_Of_Birth': ['15/05/2002', '17/05/2002', '16/02/1999', '25/09/1998', '11/05/2002', '15/09/1997'],
    'age': [12, 12, 13, 13, 14, 12],
    'height': [173, 192, 186, 167, 151, 159],
    'weight': [35, 32, 33, 30, 31, 32],
    'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']
}

df = pd.DataFrame(data, index=['S1', 'S2', 'S3', 'S4', 'S5', 'S6'])
```

```python
# 1. Write a Pandas program to split the following dataframe into groups based on school code. Also check the type of GroupBy object.

groups = df.groupby('school') # Split DataFrame into groups based on school code

# Print each group
for name, group in groups:
    print(f"Group: {name}")
    print(group, "\n")

print("Type of GroupBy object:", type(groups)) # Check the type of GroupBy object
```

```python
# 2. Write a Pandas program to split the following dataframe by school code and get mean, min, and max value of age for each school.

result = df.groupby('school')['age'].agg(['mean', 'min', 'max']) # Group by school code and calculate mean, min, and max of age
print(result)
```

```python
# 3. Write a Pandas program to split the following given dataframe into groups based on school code and class. Go to the editor

groups = df.groupby(['school', 'class'])

for name, group in groups:
    print(f"Group: {name}")
    print(group)
    print()
```

```python
"""4. Write a Pandas program to split the following given dataframe into groups based on single column and multiple columns.
Find the size of the grouped data."""

# Group by single column (school)
print("Size of groups by school:")
print(df.groupby('school').size())
print()

# Group by multiple columns (school and class)
print("Size of groups by school and class:")
print(df.groupby(['school', 'class']).size())
```

```python
"""
5. Write a Pandas program to split a dataset to group by two columns and count by each row.
    ord_no  purch_amt    ord_date  customer_id  salesman_id
0    70001     150.50  2012-10-05         3005         5002
1    70009     270.65  2012-09-10         3001         5005
2    70002      65.26  2012-10-05         3002         5001
3    70004     110.50  2012-08-17         3009         5003
4    70007     948.50  2012-09-10         3005         5002
5    70005    2400.60  2012-07-27         3007         5001
6    70008    5760.00  2012-09-10         3002         5001
7    70010    1983.43  2012-10-10         3004         5006
8    70003    2480.40  2012-10-10         3009         5003
9    70012     250.45  2012-06-27         3008         5002
10   70011      75.29  2012-08-17         3003         5007
11   70013    3045.60  2012-04-25         3002         5001
"""

import pandas as pd

data = {
    'ord_no': [70001, 70009, 70002, 70004, 70007, 70005, 70008, 70010, 70003, 70012, 70011, 70013],
    'purch_amt': [150.50, 270.65, 65.26, 110.50, 948.50, 2400.60, 5760.00, 1983.43, 2480.40, 250.45, 75.29, 3045.60],
    'ord_date': ['2012-10-05', '2012-09-10', '2012-10-05', '2012-08-17', '2012-09-10', '2012-07-27', '2012-09-10', '2012-10-10', '2012-10-10', '2012-06-27', '2012-08-17', '2012-04-25'],
    'customer_id': [3005, 3001, 3002, 3009, 3005, 3007, 3002, 3004, 3009, 3008, 3003, 3002],
    'salesman_id': [5002, 5005, 5001, 5003, 5002, 5001, 5001, 5006, 5003, 5002, 5007, 5001]
}
df = pd.DataFrame(data)

# Group by customer_id and salesman_id
result = df.groupby(['customer_id', 'salesman_id']).size().reset_index(name='count') #and count rows

print(result)
```

```python
"""
6. Write a Pandas program to split a dataset to group by two columns and then sort the aggregated results within the groups.
In the following dataset group on 'customer_id', 'salesman_id' and then sort sum of purch_amt within the groups.
Test Data:
    ord_no  purch_amt    ord_date  customer_id  salesman_id
0    70001     150.50  2012-10-05         3005         5002
1    70009     270.65  2012-09-10         3001         5005
2    70002      65.26  2012-10-05         3002         5001
3    70004     110.50  2012-08-17         3009         5003
4    70007     948.50  2012-09-10         3005         5002
5    70005    2400.60  2012-07-27         3007         5001
6    70008    5760.00  2012-09-10         3002         5001
7    70010    1983.43  2012-10-10         3004         5006
8    70003    2480.40  2012-10-10         3009         5003
9    70012     250.45  2012-06-27         3008         5002
10   70011      75.29  2012-08-17         3003         5007
11   70013    3045.60  2012-04-25         3002         5001
"""

# Group by customer_id and salesman_id, sum purch_amt
result = df.groupby(['customer_id', 'salesman_id'])['purch_amt'].sum().reset_index()
result = result.sort_values(['customer_id', 'purch_amt'], ascending=[True, False]) #sort by sum within groups
print(result)
```

```python
"""
7. Write a Pandas program to split the following dataset using group by on first column and aggregate over multiple lists on second column. Go to the editor
Test Data:
  student_id         marks
0       S001  [88, 89, 90]
1       S001  [78, 81, 60]
2       S002  [84, 83, 91]
3       S002  [84, 88, 91]
4       S003  [90, 89, 92]
5       S003  [88, 59, 90]
Output:
student_id
S001 [83.0, 85.0, 75.0]
S002 [84.0, 85.5, 91.0]
S003 [89.0, 74.0, 91.0]
"""

import pandas as pd
import numpy as np

data = {
    'student_id': ['S001', 'S001', 'S002', 'S002', 'S003', 'S003'],
    'marks': [[88, 89, 90], [78, 81, 60], [84, 83, 91], [84, 88, 91], [90, 89, 92], [88, 59, 90]]
}
df = pd.DataFrame(data)

# Group by student_id and calculate mean of each list element
result = df.groupby('student_id')['marks'].apply(lambda x: np.mean(x.tolist(), axis=0)).reset_index()
result['marks'] = result['marks'].apply(lambda x: [round(i, 1) for i in x])

print(result)
```

```python
"""
1. Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.
Sample Financial data (fdata.csv):
Date,Open,High,Low,Close
10-03-16,774.25,776.065002,769.5,772.559998
10-04-16,776.030029,778.710022,772.890015,776.429993
10-05-16,779.309998,782.070007,775.650024,776.469971
10-06-16,779,780.47998,775.539978,776.859985
10-07-16,779.659973,779.659973,770.75,775.080017
"""

import matplotlib.pyplot as plt
import pandas as pd

# Sample financial data
data = {
    'Date': ['10-03-16', '10-04-16', '10-05-16', '10-06-16', '10-07-16'],
    'Open': [774.25, 776.030029, 779.309998, 779, 779.659973],
    'High': [776.065002, 778.710022, 782.070007, 780.47998, 779.659973],
    'Low': [769.5, 772.890015, 775.650024, 775.539978, 770.75],
    'Close': [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]
}
df = pd.DataFrame(data)

# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y')

# Plot line charts
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Open'], label='Open')
plt.plot(df['Date'], df['High'], label='High')
plt.plot(df['Date'], df['Low'], label='Low')
plt.plot(df['Date'], df['Close'], label='Close')
plt.title('Alphabet Inc. Financial Data (Oct 3-7, 2016)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
"""
3. Write a Python program to plot two or more lines with legends, different widths and colors.
Matplotlib Basic: Plot two or more lines with legends, different widths and colors
"""

import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Plot lines
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Sine', color='blue', linewidth=2)
plt.plot(x, y2, label='Cosine', color='red', linewidth=3)
plt.plot(x, y3, label='Tangent', color='green', linewidth=1.5)
plt.title('Multiple Lines with Different Widths and Colors')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend() #
plt.grid(True)
plt.tight_layout()

plt.savefig('multiple_lines.png')
```

```python
# 4. Write a Python program to plot two or more lines and set the line markers. Go to the editor

# Plot lines with markers
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Sine', marker='o', color='blue', linestyle='-')
plt.plot(x, y2, label='Cosine', marker='s', color='red', linestyle='--')
plt.title('Lines with Markers')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig('lines_with_markers.png')
```
