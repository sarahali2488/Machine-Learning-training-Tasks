```
# load the iris dataset
from sklearn.datasets import load_iris
from sklearn import tree

X, y = load_iris(return_X_y=True) #X features, y labels / class
```

```
X #features are 4 values to each sample height,width,...
```

```
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```
y #labels of images 3 classees for 3 flowers' types
```

```
#Calling Decision tree classifier
clf = tree.DecisionTreeClassifier()

# Train the DT on X and y data
clf = clf.fit(X, y)
```

```
tree.plot_tree(clf) #To show the tree drawn above
```

```
# store the feature matrix (X) and response vector (y)
#X = iris.data
#y = iris.target
```

```
# splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```
# training the model on training set
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
results=gnb.fit(X_train, y_train)
results
```

```
# making predictions on the testing set
y_pred = gnb.predict(X_test)
y_pred
```

```
# comparing actual response values (y_test) with predicted response values (y_pred)
from sklearn import metrics
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
```

To increase accuracy I reduced the percentage of testing data and increased the percentage of training data
