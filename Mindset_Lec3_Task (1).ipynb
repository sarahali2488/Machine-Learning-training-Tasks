creat a new column called yesterday_temp put on it the value of tempreture column - 1 and group by the data based on the city and get the mean of the yesterday_temp.

```python
import pandas as pd

df = pd.read_csv("/content/weather_by_cities.csv")
df.head()
```

```python
# Creating new column yesterday_temp
df['yesterday_temp'] = df['temperature'] - 1
print("Table of temperatures: \n", df)

# Groupping by city and get mean of yesterday_temp
result = df.groupby('city')['yesterday_temp']
yes_temp_mean = result.mean()
print("Yesterday temp mean for cities: \n", yes_temp_mean)
```
